```python
import pandas as pd

# This imports the csv file into memory as a dataframe
# In this case the income data for 50 States for several years
df = pd.read_csv("income.csv")

# Show just the tope 5 rows of imported data
df.head(5)
```

```python
# You could also see just the last rows
df.tail(n = 5)
```

```python
# .describe()
# It gives a quick look at the scale, skew, and range of numeric data.

# Default pencentiles are 25, 50, 75 but changing here
df.describe(percentiles=[0.3, 0.5, 0.7])
```

```python
# It might be easiest to view the data by transposing with .T
df.describe().T
```

```python
# Let's clean up the dataframe

# This is the current dataframe:
df.head()
```

```python
# Remove (known as drop) column named "Index" to prepare for plotting using .drop()
# Axis number (0 for rows and 1 for columns.)
df.drop('Index', axis=1).head(5)
# Alternately, can do: df = df.drop(columns=['column_nameA', 'column_nameB'])

# This will not actually update the dataframe unless it is assigned to a variable, reassigned to itself, or has inplace=True
# Like this:
#df = df.drop('Index', axis=1)
# or
# x = df.drop('Index', axis=1)
# or
df.drop('Index', axis=1, inplace=True)

# Let's check and see if the "Index" column is gone
df.head(5)
# x.head()
```

```python
# The index according to Pandas dataframe are the numbers on the Left most row (0...4)


# Set index column to 'State"
df = df.set_index('State')

# See what changed:
df.head()
```

```python
# Import matplotlib for plotting
# import matplotlib.pyplot
import matplotlib.pyplot as plt

df.plot()
```

```python
# That is too busy - let's simplify to just a few

df.sample(n = 3).plot()
# Obviously, this is wrong - the Years should be on the x-axis and the Labels should be the States 
```

```python
# We need to transpose to get the States across the top for the labels and go by year on the x-axis

# Transpose using "T"
df.sample(n=3).T.head(5)
```

```python
# Putting it all together:

# df.sample(n = 3).T.plot(kind='line')
df.sample(n = 3).T.plot()
```

```python
# Show just Florida
df.T.loc[:,['Florida']].plot(kind='bar')

# Using index (Florida is index position 9 - index begins at 0):
# df.T.iloc[:,9].plot(kind='bar')
```
