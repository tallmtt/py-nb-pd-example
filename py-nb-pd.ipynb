```python
import pandas as pd

# This imports the csv file into memory as a dataframe
# In this case the income data for 50 States for several years
df = pd.read_csv("income.csv")

# Show just the tope 5 rows of imported data
# Note that it might wrap around indicated by "\"
df.head(5)
```

```python
# You could also see just the last rows
df.tail(n = 5)
```

```python
# .describe()
# It gives a quick look at the scale, skew, and range of numeric data.

# Default pencentiles are 25, 50, 75 but changing here
df.describe(percentiles=[0.3, 0.5, 0.7])
```

```python
# It might be easiest to view the data by transposing with .T
df.describe().T
```

```python
df.info(show_counts=True, memory_usage=True, verbose=True)
```

```python
# Get the number of rows and columns
# I am using print here to show what is going on
df.shape
print(df.shape)
# Get all columns and column names
print(df.columns)
# Convert to a list
list(df.columns)
```

```python
# Check for missing values in pandas with .isnull()
# Here there is no missing data:
df.isnull().sum()
```

```python
# Isolating one column using [ ]
# Here restricted to just the top 5 rows to keep it simple
df['Y2002'].head(5)
```

```python
# Isolating two or more columns using [[ ]] 
df[['Y2002', 'Y2003']].head(5)
```

```python
# Isolating one row using [ ] 
df[df.index==1]
```

```python
# Isolating two or more rows using [ ] 
df[df.index.isin(range(7,11))]
```

```python
# Select first 2 rows
df.iloc[:2]
```

```python
# Select columns by name (and of course only the 5 first rows)
df.loc[:,["Y2002","Y2015"]].head(5)
```

```python
# Get the mode, mean, or median
# Assign last command to a new dataframe named df2
df2 = df.loc[:,["Y2002","Y2015"]].head(5)

# Get the mean of df2 - or .mode() or .median()
df2.mean()

# In fact, all of these work: .min(), .max(), .mean(), .median(), .sum(), .mode()
```

```python
# Remove column named "Index" to prepare for plotting using .drop()
print(df.head())

# Axis number (0 for rows and 1 for columns.)
print(df.drop('Index', axis=1).head(5))
# Alternately, can do: df = df.drop(columns=['column_nameA', 'column_nameB'])

# This will not actually update the dataframe unless it is assigned to a variable, reassigned to itself, or has inplace=True
df.drop('Index', axis=1, inplace=True)

# Let's check and see if the "Index" column is gone
df.head(5)
```

```python
# Import matplotlib for plotting
# import matplotlib.pyplot
import matplotlib.pyplot as plt

# Set index column to 'State"
df = df.set_index('State')
```

```python
# df.head(3).plot()
# df.plot()
df.sample(n = 3).plot()
# Obviously, this is wrong - the Years should be on the x-axis and the Labels should be the States 
```
